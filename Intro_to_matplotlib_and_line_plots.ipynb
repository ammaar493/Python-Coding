There are two ways to select rows:<br>
- df.loc[label] # filters out rows by labels of the index/column
- df.iloc[index] # filters out rows by the positions of the index/column

## Line plots
Line plot is a type of plot used for contineus data. It is constructed by plotting points on graph and connecting them with a stright line segments.

***Question:*** Plot a line graph of immigrants from Haiti.

```python
# First extract the data of Haiti of all years
haiti = df_can.loc['Haiti', years]
haiti.head() # Here we will see first five years
```

```python
# Now plot that data
haiti.plot()
```

Here we can see a peek around 2010. In 2010 there was a catastrophic earthquake in Haiti, as part of Canada's humanitarian effort, the government of Canada stepped up and accepting refugees from Haiti.

Pandas automatically polpulated the x-axis with index values and y-axis with the column values. See how the years are displayed because years are of string type. 

Let's change somethings:<br>
- Change years datatype to interger
- Add plot title
- Add y label
- Add x label

```python
haiti.index = haiti.index.map(int)
haiti.plot(kind = 'line')

plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')

plt.show()
```

Let's add text to plot

```python
haiti.plot(kind = 'line')

plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')


plt.text(2001,6300, '2010 Earthquake') # plt.text(x,y, label)

plt.show()
```
